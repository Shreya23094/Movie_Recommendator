```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(2)
```

```python
credits.head(2)
```

```python
mm=movies.merge(credits,on='title')
```

```python
mm.head(1)
```

```python
#genre, id, keywords, title, overview, cast, crew
df=mm[['movie_id', 'title', 'overview','genres', 'keywords', 'cast', 'crew']]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[0].keywords
```

```python
import ast 
def convert(obj):
    x=[]
    obj=ast.literal_eval(obj)
    for i in obj:
        x.append(i['name'])
    return x
```

```python
df['genres']=df['genres'].apply(convert)
df['keywords']=df['keywords'].apply(convert)
```

```python
df.head()
```

```python
def convert_3(obj):
    x=[]
    obj=ast.literal_eval(obj)
    count=0
    for i in obj:
        if count!=3:
            x.append(i['name'])
            count+=1
        else:
            break
    return x
```

```python
df['cast']=df['cast'].apply(convert_3)
```

```python
a=ast.literal_eval(df.crew[0])
a[6]
```

```python
def fetch_director(obj):
    x=[]
    obj=ast.literal_eval(obj)
    for i in obj:
        if i['job']=='Director':
            x.append(i['name'])
            break
    return x
```

```python
df.crew=df.crew.apply(fetch_director)
```

```python
df.head()
```

```python
df['overview'][0]
```

```python
df['overview']=df['overview'].apply(lambda x: x.split())
```

```python
df['genres']=df['genres'].apply(lambda x : [i.replace(' ','') for i in x])
df['cast']=df['cast'].apply(lambda x : [i.replace(' ','') for i in x])
df['crew']=df['crew'].apply(lambda x : [i.replace(' ','') for i in x])
df['keywords']=df['keywords'].apply(lambda x : [i.replace(' ','') for i in x])
```

```python
df.head()
```

```python
df['tags']=df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
new_df=df[['movie_id','title','tags']]
new_df['tags']=new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

## Info about the next step:
We have our tags that will help us define for model for our movie recoommendation but the issue here is that we cannot directly recommend movies by this so here we can convert these tags into vectors using:
#### CountVectorizer:
When we convert tags into vectors, it will help us to identify which typed word is close to what we can recommend from the tags

##### Side note: Stop words
stop words are those words which we may not need in a model and are there to stucturize a sentence like a, an, with, to, etc.
we need to remove those in our model.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

### PorterStemmer:
we need to remove the prefixes and suffixes from the words so that we get accurate results for which we use stemming. Stemming is a text processing method, used to examine words by analyzing prefix and suffix.

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem_func(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem_func)
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
cv.get_feature_names_out(vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
new_df[new_df['title']=='Avatar'].index[0]
```

```python
def recommend(movie):
    mindex=new_df[new_df['title']==movie].index[0]
    dist=similarity[mindex]
    mlist=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in mlist:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
